```python
# To keep things organized, we are going to make a function that performs that cleaning process.
# Remember, functions are blocks of code within a script or algorithm that perform a specific task. 
# There are four basic parts to a function:

                                         # Name
                                         # Parameters
                                         # Code block
                                         # Return value
```

```python
x = 'global value'

def foo():
    x = 'local value'
    print(x)

foo()
print(x)
```

```python
# When passing mutable objects like a dict or list as parameters to a function, 
# the function can change the values inside the object.

my_list = [1,2,3]
def append_four(x):
    x.append(4)
append_four(my_list)
print(my_list)
```

```python
# It changed my_list to [1,2,3,4], even outside the function. 
#To leave the original object unchanged, we need to make a copy. 
#For lists, we make a copy with the list function; for dicts, 
# we make a copy with the dict function, like so:

# new_list = list(old_list)
# new_dict = dict(old_dict)
```

```python
# Lambda Functions: are written in one line and automatically return a value without using the return keyword. 
# Lambda functions have no name and are also known as "anonymous functions."
```

```python
square = lambda x: x * x
square(5)
```
